## Exercise 1

Plot the logistic function 


$$f(x) = \frac{e^x}{1 + e^x}$$

Between -7 and 7

````python
a =-7
b = 7
n = 50

x = np.linspace(a, b, n)
y = np.e**x / (1+np.e**x)

plt.plot(x, y)
plt.show()
````

## Exercise 2 

Given this code:
```python
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.plot (X, Y-1, color='blue', alpha=1.00)
plt.show()
```

Fill in the plot with colors:

````python
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.figure(figsize=(8,6))

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.fill_between(X,1,Y+1, color='blue', alpha=0.25)

plt.plot (X, Y-1, color='blue', alpha=1.00)
plt.fill_between(X,-1,Y-1, color='blue', alpha=0.25, where=Y-1>-1)
plt.fill_between(X,-1,Y-1, color='red', alpha=0.25, where=Y-1<-1)

plt.ylim(-2.5,2.5)
plt.xticks([])
plt.yticks([])

plt.grid(False)
plt.show()
````

## Exercise 3

Create a plot for the functions $\sin(x)$ y $\exp(x)$, between $-\pi$ y $\pi$, with different styles for each function. The plot has to include a legend and the name of the axes:
* xlabel = '$\alpha$ (rad)'
* ylabel = 'f'

No grid and fill the space between the two plots.

````python
x=np.linspace(-np.pi, np.pi, 500)
y1 = np.sin(x)
y2 = np.exp(x)

fig, ax = plt.subplots(figsize=(7,5))

plt.plot(x, y1, c='b', ls='-.', label='sin(x)')
plt.plot(x, y2, c='r', ls=':', label='exp(x)'), 

plt.fill_between(x, y1, y2, color='orange' , alpha=0.25)

plt.legend()
plt.grid(False)
ax.set_xlabel(r'$\alpha(rad)$', fontsize=16)
ax.set_ylabel("f", fontsize=16)
plt.show()
````
