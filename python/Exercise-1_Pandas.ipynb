```python
tips = pd.read_csv('tips.csv')
tips.head()
```

### Exercise 1

Obtener el porcentaje de propina y analizar la dependencia con las variables sexo y fumadores (sex y smoker). ¿Se puede ver alguna diferencia de comporamiento entre hombres/mujeres, y si son o no fumadores?

```python
tips['%tip'] = np.round(100 * tips['tip'] / tips['total_bill'], decimals=2)
tips.head()
```

```python
tips_group = tips.groupby(['sex', 'smoker'])
['%tip'].describe()
```

```python
def peak_to_peak(s):
    return s.max() - s.min()

def rango_normal(s):
    return 4*s.std()
```

```python
tips_group['%tip'].agg(['mean', ('std dev','std'), 'count', 'min', 'max', ('range', peak_to_peak),('range (95%)',rango_normal)])
```
